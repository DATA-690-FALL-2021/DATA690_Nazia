## 6.Cleanse the data to make the year, population, cost, cost per person numerical type 

```python
# remove ",","$" from the dataframe.
df = df.apply(lambda x: x.str.replace(',', ''))
df = df.apply(lambda x: x.str.replace('$', ''))

#remove "*" from the "Census Year" column
df['Census Year'] = df['Census Year'].apply(lambda x: x.replace('*', ''))
```

```python
#function for removing cents and convert it to dollar
def cent_to_dollar(a):
  if 'cents' in a:
    a=a.replace("cents","")
    a=(float(a))/100
    return a
  else:
    return a
  
```

```python
df["Average Cost Per Person"]=df["Average Cost Per Person"].apply(cent_to_dollar)
```

```python
df
```

```python
#function for removing "Billion" and add necessary 0 instead of it
def remove_billion(a):
  if "Billion" in a:
    a=a.replace("Billion","")
    a=(float(a))*1000000000
    return a
  else:
    return a
```

```python
df["Census Cost"]=df["Census Cost"].apply(remove_billion)
```

```python
df
```
