## 15.Calculate the min, max, and average in-state tution of the above four institutions

```python
# create new datafrme for finding min, max & avg
new_df2 = pd.concat([a, b, c, d], axis=0)
new_df2
```

```python
# find minimin in-state tuition fee 
min_tuition = new_df2['TUITIONFEE_IN'].min()
print("Minimum in-state tuition fee is", min_tuition)
```

```python
# find maximum in-state tuition fee
max_tuition = new_df2['TUITIONFEE_IN'].max()
print("Maximum in-state tuition fee is", max_tuition)
```

```python
# find average in-state tuition fee
avg_tuition = new_df2['TUITIONFEE_IN'].mean()
print("Average in-state tuition fee is", avg_tuition)
```
